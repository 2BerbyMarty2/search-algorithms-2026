```python
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
def get_coords(node_id):
  x = node_id // 6
  y = node_id % 6
  return (x,y)
```

```python
get_coords(31)
```

```python
def setup_maze():
  start_id = random.randint(0, 11)
  goal_id = random.randint(24, 35)

  all_nodes = list(range(36))
  valid_barriers = [n for n in all_nodes if n!= start_id and n!= goal_id]
  barriers = random.sample(valid_barriers, 4)

  return {
      'start': get_coords(start_id),
      'goal': get_coords(goal_id),
      'barriers': [get_coords(b) for b in barriers]
  }
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def draw_maze(maze_coordinates):
  fig, ax = plt.subplots(figsize=(5, 5))

  # 1. Define 6 boxes (0, 1, 2, 3, 4, 5)
  labels = range(6)
  centers = np.arange(0.5, 6.0)

  # 2. Place labels in the center of the boxes
  ax.set_xticks(centers)
  ax.set_yticks(centers)
  ax.set_xticklabels(labels)
  ax.set_yticklabels(labels)

  # 3. Draw grid lines at the box edges (0 through 6)
  ax.set_xticks(range(7), minor=True)
  ax.set_yticks(range(7), minor=True)
  ax.grid(True, which='minor', color='black', linestyle='-')

  # 4. Clean up: hide tick marks and move X-axis to top
  ax.tick_params(which='major', length=0)
  ax.xaxis.tick_top()

  # 5. Invert Y and set limits to show exactly 6 boxes
  ax.invert_yaxis()
  ax.set_xlim(0, 6)
  ax.set_ylim(6, 0)

  # Example: Add a green box at (0,0)
  ax.add_patch(patches.Rectangle(maze_coordinates['start'], 1, 1, color='green', alpha=0.6, label='Start Box'))
  ax.add_patch(patches.Rectangle(maze_coordinates['goal'], 1, 1, color='red', alpha=0.6, label='Goal Box'))
  for barrier_coords in maze_coordinates['barriers']:
    ax.add_patch(patches.Rectangle(barrier_coords, 1, 1, color='Black', alpha=0.6, label='Black Box'))

  plt.show()
```

```python
draw_maze(setup_maze())
```
